```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
fig,ax = plt.subplots(nrows = 1,ncols=1, figsize=(10,9.5))
k = 10
n = 100
y_data = np.arange(n)
(np.random.shuffle(y_data))
ax.scatter([i for i in range (n)], y_data,s=25)
ax.grid(b = 'True',alpha = 0.35)
ax.set_xlim(-0.65,100.1)
ax.set_ylim(-0.65,100.1)
ax.set_title (r'Permutation without structure $\Pi$',fontsize = 35,pad=-5,color='b')
ax.tick_params(direction='in')
ax.set_xticklabels([])
ax.set_yticklabels([])
plt.savefig("general.png", bbox_inches = 'tight',
    pad_inches = 0.1)
```

```python
fig,ax = plt.subplots(nrows = 1,ncols=1, figsize=(10,9.5))
k = 25
n = 100
y_data = np.arange(n)
idx    = np.random.permutation(y_data)[:k]
y_data[idx] = np.random.permutation(idx)
ax.scatter([n - i for i in range (n)], y_data,s=25,marker="o")
ax.grid(b = 'True',alpha = 0.35)
ax.set_xlim(-0.25,100.25)
ax.set_ylim(-0.00,100.25)
#ax.set_title (r'Sparse permutation $\Pi_o$',fontsize = 35,pad = -5,color='b')
ax.tick_params(direction='in')
#ax.set_xticklabels([])
#ax.set_yticklabels([])
ax.tick_params(size=10,direction='in',labelsize=25)
plt.savefig("sparse.pdf", bbox_inches = 'tight',
    pad_inches = 0.1)
```

```python
fig,ax = plt.subplots(nrows = 1,ncols=1, figsize=(10,9.25))
r = 25
n = 100
y_data = np.arange(r)
ax.scatter([n - i for i in range (n)],[j*r + np.random.permutation(y_data) for j in range (int(n/r))],s=25)
ax.set_xlim(-0.25,100.25)
ax.set_ylim(-0.00,100.25)
#ax.set_title (r'$r$-local permutation $\Pi_r$',fontsize = 35,pad = -5,color='b')
ax.set_xticks(range(0,n+1,r))
ax.set_yticks(range(0,n+1,r))
ax.grid(b = 'True',alpha = 1,linewidth = 2)
ax.tick_params(size=10,direction='in',labelsize=25)
plt.savefig("r_local.pdf", bbox_inches = 'tight',
    pad_inches = 0.1)
```

```python
fig,ax = plt.subplots(nrows = 1,ncols=1, figsize=(10,9.25))
r = 20
n = 100
y_data = np.arange(r)
blk = np.arange(int(n/r))
np.random.shuffle(blk)
ax.scatter(range(0,n),[idx*r + np.random.permutation(y_data) for idx in blk],s=25)
ax.set_xlim(-0.25,100.25)
ax.set_ylim(-0.00,100.25)
ax.set_title (r'Clustered graph matching $\Pi_{pq}$',fontsize = 35,pad = -5,color='b')
#ax.set_xticks(range(0,n+1,r))
#ax.set_yticks(range(0,n+1,r))
ax.grid(b = 'True',alpha = 1,linewidth = 2)
ax.tick_params(size=10,direction='in',labelsize=25)
plt.savefig("clustered_graph_matching.png", bbox_inches = 'tight',
    pad_inches = 0.1)
```

```python
def get_permutation (n,num_assigned):
    P = np.eye(n)
    idx = np.random.permutation(n)[0:n-num_assigned]
    P[idx,idx] = 0
    P[idx,np.random.permutation(idx)] = 1    
    return P
```

```python
n = 100
num_assigned = 50
P = get_permutation(n,num_assigned)
fig = plt.figure()
ax1 = fig.add_subplot(121)
ax1.imshow(P)
```
