```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
fig,ax = plt.subplots(nrows = 1,ncols=1, figsize=(10,9.5))
k = 10
n = 100
y_data = np.arange(n)
(np.random.shuffle(y_data))
ax.scatter([i for i in range (n)], y_data,s=25)
ax.grid(b = 'True',alpha = 0.35)
ax.set_xlim(-0.65,100.1)
ax.set_ylim(-0.65,100.1)
ax.set_title (r'Permutation without structure $\Pi$',fontsize = 35,pad=-5,color='b')
ax.tick_params(direction='in')
ax.set_xticklabels([])
ax.set_yticklabels([])
plt.savefig("general.png", bbox_inches = 'tight',
    pad_inches = 0.1)
```

```python
[fig,ax]  = plt.subplots(1, 2, figsize = (7,3))
k = 15
n = 30
y_data = np.arange(n)
idx    = np.random.permutation(y_data)[:k]
y_data[idx] = np.random.permutation(idx)
ax[0].scatter([n - i for i in range (n)], y_data,s=11,marker="o",color='black')
ax[0].grid(b = 'True',alpha = 0.35)
ax[0].set_xlim(-0.35,30.35)
ax[0].set_ylim(-0.35,30.35)
#ax.set_title (r'Sparse permutation $\Pi_o$',fontsize = 35,pad = -5,color='b')
ax[0].tick_params(direction='in')
ax[0].set_xticks([0,10,20,30])
ax[0].set_yticks([0,10,20,30])
#ax.set_xticklabels([])
#ax.set_yticklabels([])
ax[0].tick_params(size=5,direction='in',labelsize=11)
r = 10
n = 30
y_data = np.arange(r)
ax[1].scatter([n - i for i in range (n)],[j*r + np.random.permutation(y_data) for j in range (int(n/r))],s=11,color='black')
ax[1].set_xlim(-0.35,30.35)
ax[1].set_ylim(-0.35,30.35)
#ax.set_title (r'$r$-local permutation $\Pi_r$',fontsize = 35,pad = -5,color='b')
ax[1].set_xticks(range(0,n+1,r))
ax[1].set_yticks(range(0,n+1,r))
ax[1].grid(b = 'True',alpha = 1,linewidth = 2)
ax[1].tick_params(size=5,direction='in',labelsize=11)
#plt.savefig("r_local.pdf", bbox_inches = 'tight',
#    pad_inches = 0.1)
plt.savefig("combined.pdf", bbox_inches = 'tight',
    pad_inches = 0.1)
```

```python
def get_permutation (n,num_assigned):
    P = np.eye(n)
    idx = np.random.permutation(n)[0:n-num_assigned]
    P[idx,idx] = 0
    P[idx,np.random.permutation(idx)] = 1    
    return P
```

```python
n = 100
num_assigned = 50
P = get_permutation(n,num_assigned)
fig = plt.figure()
ax1 = fig.add_subplot(121)
ax1.imshow(P)
```
